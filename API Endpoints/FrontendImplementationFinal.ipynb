```python
from fastapi import FastAPI
from pydantic import BaseModel
from fastapi.responses import JSONResponse
from fastapi.middleware.cors import CORSMiddleware
import os
import uuid
import requests
from moviepy.editor import VideoFileClip, concatenate_videoclips
from pyngrok import ngrok
import cloudinary
from cloudinary import uploader
import nest_asyncio
import uvicorn
from threading import Thread

# === Setup ===
SAVE_DIR = "./video_clips"
FINAL_OUTPUT_TEMPLATE = "./final_video_{id}.mp4"
os.makedirs(SAVE_DIR, exist_ok=True)

# === Cloudinary Config ===
cloudinary.config(
    cloud_name="dxjjpowf1",
    api_key="276811499447944",
    api_secret="Eeb9f4MTWcmsPCxXsujuv46nOXs",
    secure=True
)

# === FastAPI App ===
app = FastAPI()

# === Enable CORS ===
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# === Request Model ===
class VideoRequest(BaseModel):
    tts_url: str
    clip_url: str
    ref_video: str
    ref_audio: str
    script: str

# === File Download Helper ===
def download_file(url, save_path):
    try:
        response = requests.get(url, stream=True, timeout=30)
        response.raise_for_status()
        with open(save_path, 'wb') as f:
            for chunk in response.iter_content(8192):
                f.write(chunk)
        return True
    except Exception as e:
        print(f"Download failed: {str(e)}")
        return False

# === Main Video Generation Logic ===
def generate_final_video(tts_url, clip_url, ref_video_url, ref_audio_url, final_output_path, script):
    try:
        tts_data = {
            "audio_url": ref_audio_url,
            "text": f"Tell me about {script} in strictly less than 50 words. No unnecessary explanations"
        }
        tts_response = requests.post(tts_url, json=tts_data, timeout=300)
        if tts_response.status_code != 200:
            print("TTS error")
            return None

        audio_urls = tts_response.json().get("video_files", [])
        clip_urls = []

        for audio_url in audio_urls:
            clip_data = {
                "video_url": ref_video_url,
                "audio_url": audio_url
            }
            clip_response = requests.post(clip_url, json=clip_data, timeout=1000)
            if clip_response.status_code == 200:
                clip_urls.append(clip_response.json().get("video_url"))

        clip_paths = []
        for url in clip_urls:
            filename = os.path.basename(url)
            save_path = os.path.join(SAVE_DIR, filename)
            if download_file(url, save_path):
                clip_paths.append(save_path)

        clips = []
        for path in clip_paths:
            try:
                clips.append(VideoFileClip(path))
            except Exception as e:
                print(f"Error loading clip {path}: {e}")
                continue

        if not clips:
            print("No clips to merge")
            return None

        final_clip = concatenate_videoclips(clips, method="chain")
        final_clip.write_videofile(final_output_path, codec="libx264")

        for clip in clips:
            clip.close()
        for path in clip_paths:
            os.remove(path)
        os.rmdir(SAVE_DIR)

        return final_output_path

    except Exception as e:
        print(f"Error in generate_final_video: {e}")
        return None

# === POST Endpoint ===
@app.post("/create_video")
def create_video(request: VideoRequest):
    unique_id = str(uuid.uuid4())[:8]
    output_path = FINAL_OUTPUT_TEMPLATE.format(id=unique_id)

    print("üîÑ Starting video creation process...")
    result = generate_final_video(
        request.tts_url,
        request.clip_url,
        request.ref_video,
        request.ref_audio,
        output_path,
        request.script
    )

    if not result:
        return JSONResponse(status_code=500, content={"error": "Video creation failed"})

    print("‚òÅÔ∏è Uploading video to Cloudinary...")
    try:
        upload_result = uploader.upload(
            output_path,
            resource_type="video",
            folder="generated_videos",
            timeout=300
        )
        os.remove(output_path)
        return {
            "status": "success",
            "cloudinary_url": upload_result["secure_url"]
        }
    except Exception as e:
        return JSONResponse(status_code=500, content={"error": f"Upload failed: {str(e)}"})

# === Ngrok Tunnel ===
def run_ngrok():
    ngrok.set_auth_token("2vtiguRHM1mbsTXAsfj8w9IZbSi_6r34VwkFY6b3BKtGQ8F9s")
    public_url = ngrok.connect(8000)
    print(f"\nüöÄ Public URL: {public_url}\n")

def start():
    Thread(target=run_ngrok).start()

# === Main Entry ===
if __name__ == "__main__":
    start()
    nest_asyncio.apply()
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
